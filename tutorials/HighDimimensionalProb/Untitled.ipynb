```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
def sim(n, d):
    return np.random.normal(0,1,(n,d))
```

```python
ds = [2,5,10,25,50,100,1000,2000,5000]
```

```python
datas = []
for d in ds:
    datas.append(sim(100,d))
```

```python
radii = [np.linalg.norm(data, axis=1).mean() for data in datas]
plt.scatter(ds,radii)
plt.plot(ds,np.sqrt(ds))
plt.title(f'N(0,1) Radius vs dimension')
plt.xlabel('Dimension')
plt.ylabel('Radius')
plt.show()
```
