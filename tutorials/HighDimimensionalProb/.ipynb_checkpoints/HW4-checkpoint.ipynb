```python
import numpy as np
from scipy.sparse.linalg import svds
from itertools import product

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_context('talk')
```

Construct, for any $m,n$, a matrix $A$ whose entries are random independent, mean zero, sub-gaussian random variables.

```python
def sample_A(n,m):
    ## Sample random matrix A with standard normal 
    ## (hence sub-gaussian) entries
    return np.random.normal(0,1,size=(n,m))
```

```python
def matrix_norm(A, k=1):
    ## Returns the max eigenvalue
    ## (the 2,2 matrix norm) or all eigenvalues
    if k == 1:
        return max(svds(A, k=1, return_singular_vectors=False))
    else:
        return np.linalg.svd(A, compute_uv=False)
```

Compute $||A||$ and plot it as a function of: 
- $n$ and $m$ (separately)
- $m = n = p$ (square matrices of increasing size)
- $\mathbb{E}||A||$ and its standard deviation (over $m,n$, averaging multiple draws of $||A||$)

```python
## m,n separately
ps = [5,10,20,50,100,200,500,1000,2000]
norms_neq = [matrix_norm(sample_A(n,m)) for (n,m) in product(ps,ps)]
norms_neq = np.asarray(norms_neq).reshape((len(ps), len(ps)))
fig, ax = plt.subplots(figsize=(7, 5))
ax = sns.heatmap(norms_neq,
                 xticklabels=ps,
                 yticklabels=ps,
                square=True,
                cbar_kws={'label': '$||A||$',
                          'orientation': 'vertical'},
                cmap='Blues')
ax.invert_yaxis()
plt.xlabel('m')
plt.ylabel('n')
plt.title('$||A||$ across $m$ and $n$')
plt.tight_layout()
plt.savefig('HW3-not_equal_dims.png')
```

```python
## m=n=p
ps = [5,10,20,50,100,200,500,1000,2000]
norms_square = [matrix_norm(sample_A(p,p)) for p in ps]
plt.plot(ps, norms_square)
plt.xlabel('m,n')
plt.ylabel('$||A||$')
plt.title('$||A||$ across $m=n$')
plt.tight_layout()
plt.savefig('HW3-equal_dims.png')
```

```python
##E||A|| m,n separately
k=10
ps = [5,10,20,50,100,200,500,1000,2000]
norms_neq = [[matrix_norm(sample_A(n,m)) for (n,m) in product(ps,ps)] for _ in range(k)]
norms_neq = np.asarray(norms_neq).reshape((k, len(ps), len(ps)))
avg = np.mean(norms_neq, axis=0)
std = np.std(norms_neq, axis=0)
```

```python
fig, axs = plt.subplots(1,2,figsize=(15, 5))
axs[0] = sns.heatmap(avg,
                 xticklabels=ps,
                 yticklabels=ps,
                square=True,
                cbar_kws={'label': '$||A||$',
                          'orientation': 'vertical'},
                cmap='Blues',
                ax=axs[0])
axs[0].invert_yaxis()
axs[0].set_xlabel('m')
axs[0].set_ylabel('n')
axs[0].set_title('$\mathbb{E}||A||$')

axs[1] = sns.heatmap(std,
                 xticklabels=ps,
                 yticklabels=ps,
                square=True,
                cbar_kws={'label': 'Standard Deviation',
                          'orientation': 'vertical'},
                cmap='Blues',
                ax=axs[1])
axs[1].invert_yaxis()
axs[1].set_xlabel('m')
axs[1].set_ylabel('n')
axs[1].set_title('$||A||$ standard deviation')
plt.tight_layout()
plt.savefig('HW3-not_equal_dims-Expectation.png')
```

```python
## E||A|| m=n=p
k=10
ps = [5,10,20,50,100,200,500,1000,2000]
norms_eq = [[matrix_norm(sample_A(p,p)) for p in ps] for _ in range(k)]
norms_eq = np.asarray(norms_eq)
avg_eq = np.mean(norms_eq, axis=0)
std_eq = np.std(norms_eq, axis=0)
```

```python
fig, axs = plt.subplots(1,2,figsize=(16, 5))
axs[0].plot(ps, avg_eq)
axs[0].set_xlabel('p')
axs[0].set_ylabel('$\mathbb{E}||A||$')
axs[0].set_title('$\mathbb{E}||A||$')

axs[1].plot(ps, std_eq)
axs[1].set_xlabel('p')
axs[1].set_ylabel('$\sigma$')
axs[1].set_title('$||A||$ standard deviation')
plt.tight_layout()
plt.savefig('HW3-equal_dims-Expectation.png')
```

**Q**: Do $m,n$ need to be really large for the predictions to be reasonably accurate?

Plot all singular values of $\frac{1}{\sqrt{m}}A$ in the case where $n=m=p$ as a histogram. What do the histograms converge to as $m \rightarrow +\infty$?

```python
## Hisotgram m=n=p
k=5
ps = [5,10,20,50,100,200,500,1000]
singular_values = {p:np.asarray([matrix_norm(sample_A(p,p)/np.sqrt(p), k=-1) for _ in range(int(max(ps)*k/p))]).flatten() for p in ps}
```

```python
fig, axs = plt.subplots(2,4,figsize=(18, 7), sharex=True, sharey = True)
for (p,s),ax in zip(singular_values.items(), axs.flatten()):
    ax.hist(s, density=True)
    ax.set_title(f'm={p}')
    #ax.set_yscale('log', nonposy='clip')
    #ax.set_ylim(bottom=1)#, top=10**4)
    
fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
#plt.title('Distributions of 500 points on random lines for varying d')
plt.ylabel('n points')
plt.xlabel('Singular Values')
ax = plt.gca()
ax.yaxis.set_label_coords(-0.07, 0.5)
plt.tight_layout()
plt.savefig('HW3-hist.png')
```
