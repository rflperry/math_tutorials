# Solving the Cocktail Party Problem: PCA vs. ICA 
Problem adapted from Elements of Statistical Learning by Hastie and Friedman. Principle component analysis (PCA) is a common dimensionality reduction tool which relies on the assumption that the underlying latent components are normally distributed. Independent component analysis (ICA) relies on the assumption of non-normality and is more applicable in such situations as we will seee here

```python
import numpy as np
from sklearn.decomposition import FastICA, PCA
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(0)
n_samples = 1000
t = np.linspace(0,10,n_samples)

s1 = np.sin(3*t) # Signal 1: sinusoidal
s2 = signal.sawtooth(np.pi*t) # Signal 2: sawtooth
S = np.c_[s1, s2]
S += np.random.normal(0,0.2,S.shape)
S /= S.std(axis=0)  # Standardize data

# Mix data
A = np.array([[0.75, 0.25], [0.25, 0.75]])  # Mixing matrix
X = np.dot(S, A.T)  # Generate observations
```

The observed data are a mix of two signals (the so-called cocktail party problem).

```python
plt.subplot(2, 1, 1)
plt.title('Raw Signals')
plt.plot(S)
plt.subplot(2, 1, 2)
plt.title('Observed Signals')
plt.plot(X)
plt.tight_layout()
plt.show()
```

```python
# Compute PCA
pca = PCA(n_components=2)
S_pca = pca.fit_transform(X)  # Reconstruct signals based on orthogonal components

# Compute ICA
ica = FastICA(n_components=2)
S_ica = ica.fit_transform(X)  # Reconstruct signals
```

ICA is better able to recover the underlying signal as the data is from a non-normal process.

```python
plt.subplot(2, 1, 1)
plt.title('PCA Reconstruction')
plt.plot(S_pca)
plt.subplot(2, 1, 2)
plt.title('ICA Reconstruction')
plt.plot(S_ica)
plt.tight_layout()
plt.show()
```
